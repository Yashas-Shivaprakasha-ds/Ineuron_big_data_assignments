# Q9. How do you concatenate two Pandas DataFrames?

```python
import pandas as pd
data_01 = {
    "name": "abc",
    "skills": ["java", "python"],
    "years_of_exp" : 2
}
df1 = pd.DataFrame(data_01)
print(df1)

data_02 = {
    "name": "xyz",
    "skills": ["javascript", "python"],
    "years_of_exp" : 2.5
}
df2 = pd.DataFrame(data_02)
print(df2)

concat_res = pd.concat([df1,df2], axis = 1)
concat_res
```

# Q25. What is the difference between .loc and .iloc in Pandas?

```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv", sep=",")
# loc : It is label based, i.e, we have to specify rows and columns based on their row and column labels
# Syntax: loc[row_label, column_label]
df.loc[2, "Name"]
```

```python
# iloc: It cis integer position-based, we have to specify rows and columns by their integer position values (0-based integer position)
# Syntax: iloc[row_position, column_position]
df.iloc[2, 2]
```
